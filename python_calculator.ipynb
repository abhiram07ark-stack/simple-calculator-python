```python
num1=int(input("Enter a number"))
num2=int(input("Enter another number"))
op=input("Enter a operation from(+,-,*,/)")
def add (num1,num2):
        print(int(num1+num2)) 
def sub (num1,num2):
        print(int(num1-num2))
def multiply (num1,num2):
        print(num1*num2)
def divide (num1,num2):
        print(num1/num2)
if op=="+":
    add(num1,num2)
elif op=="-":
    sub(num1,num2)
elif op=="*":
    multiply(num1,num2)
elif op=="/":
    divide(num1,num2)
else:
    print("Invalid Operator")
```
